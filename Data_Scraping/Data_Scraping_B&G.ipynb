```python
#Get information from B&G wiki

import urllib.request, json
newline = '\n'

search_for = input()
search_for2 = search_for.replace(' ', '_')
weblink = f"https://wiki.beeldengeluid.nl/api.php?action=parse&format=json&page={search_for2}&prop=wikitext&formatversion=2"
example_wo_userinput = "https://wiki.beeldengeluid.nl/api.php?action=parse&format=json&page=Heinze_Bakker&prop=wikitext&formatversion=2"

with urllib.request.urlopen(weblink) as url:
    data = json.loads(url.read().decode())
    for element in data['parse']:
        if element == 'pageid':
            pageid_nr = data['parse'][element]
            print(pageid_nr)
        else:
            for element2 in element, data['parse'][element]:
                element2 = element2.replace(':category:','')
                element2 = element2.replace(':Category:','')
                element2 = element2.replace('=', ':')
                element2 = element2.replace('| ',';')
                element2 = element2.replace(newline, '')
                element2 = element2.replace('{', '')
                element2 = element2.replace('}', '')
                element2 = element2.replace(';', newline)
                element2 = element2.replace(',', '&')
                element2 = element2.replace(newline, ',')
                #element2 = element2.replace('[', '')
                #element2 = element2.replace(']', '')
                the_info = element2
the_list = the_info.split(',')
for element in the_list:
    if element.startswith('onderschrift'):
        the_list.remove(element)
for element in the_list:
    if element.startswith('media'):
        the_list.remove(element)
for element in the_list:
    if element.startswith('externe_info '):
        the_list.remove(element)

for element in the_list:
    if element.startswith('catalogus '):
        catalogus_option = element
        crop0 = element.find(']]')
        crop0 = int(crop0)+2
        catalogus_option = element[crop0:]
        the_list.remove(element)

for element in the_list:
    if element.startswith('programmaoverzicht '):
        programma_option = element
        crop1 = element.find(']]')
        crop1 = int(crop1)+2
        programma_option = element[crop1:]
        crop2 = programma_option.find('[[Category')
        crop2 = int(crop2)
        programma_option = programma_option[:crop2]
        programma_option = programma_option.replace(':', ';') 
        if not '.' in programma_option and '.' not in catalogus_option:
            the_list.remove(element)
            the_list.append("tekstuele informatie : ")
        elif len(programma_option) > len(catalogus_option):
            the_list.remove(element)
            programma_option = "tekstuele informatie : "+programma_option
            the_list.append(programma_option)
        else:
            the_list.remove(element)
            catalogus_option = "tekstuele informatie : "+catalogus_option
            the_list.append(catalogus_option)
for element in the_list:
    for element2 in element:
        if element2 == '[' or element2 == ']':
            element.replace(element2, '')


second_list = list()
for element in the_list:
    prep = element.split(':')
    second_list.append(prep)
column_list = list()
values_list = list()
for list_item in second_list:
    if not list_item[0] in column_list:
        column_list.append(list_item[0])
for list_item in second_list:
    for word in list_item:
        if not word in column_list:
            values_list.append(word)
column_list.remove(column_list[0])
column_list.remove(column_list[0])
values_list.remove(values_list[0])

if element_replace2 == 'tekstuele informatie : ':
    element_replace2 = f'tekstuele informatie : {max(values_list, key=len)}'


import csv

fields = column_list
    
    
# data rows of csv file  
bad_chars = ['|', '&', '<', '>', 'br']
rows = values_list 
rows.append(element_replace2)
for row in rows:
    for item in bad_chars:
        rows[5] = rows[5].replace(item, ' ')
        rows[5] = rows[5].replace('  ', ' ')
        rows[5] = rows[5].casefold()
        
rows[5] = rows[5].split(" ")
rows[5] = set(rows[5])
rows[5] = list(rows[5])
rows[5] = str(rows[5])

rows[5] = rows[5].replace(',', ',')
rows[5] = rows[5].replace("'", '')
rows[5] = rows[5].replace('[', '')
rows[5] = rows[5].replace(']', '')

rows[6] = rows[6].replace(',', ',')
rows[6] = rows[6].replace("'", '')
rows[6] = rows[6].replace('[', '')
rows[6] = rows[6].replace(']', '')

rows[8] = rows[8].replace(',', ',')
rows[8] = rows[8].replace("'", '')
rows[8] = rows[8].replace('[', '')
rows[8] = rows[8].replace(']', '')


rows[10] = rows[10].replace('&', ',')
rows[10] = rows[10].replace("'", '')
rows[10] = rows[10].replace('[', '')
rows[10] = rows[10].replace(']', '')
rows[10] = rows[10].replace(';', ':')
rows[10] = rows[10].replace('.', '. ')


print()
for row in rows:
    for item in bad_chars:
        rows[6] = rows[6].replace(item, ',')
        rows[8] = rows[8].replace(item, ',')



bad_chars2 = '[', ']', '|'

for item in bad_chars2:
    for row in rows:
        if item in row:
            rows.remove(row)
for item in bad_chars2:
    for row in rows:
        if item in row:
            rows.remove(row)
            
for row in rows:
    to_delete = 'Category'
    if to_delete in row:
        rows.remove(row)
        new_row = row.replace(to_delete, '')
        rows.append(new_row)
print(fields)
print(rows)
fields.insert(0, 'pageid')
rows.insert(0, pageid_nr)

with open('BenG_Wiki_Actors_DIRECTOUTPUT.csv', 'a', newline='') as csvfile: 
    writer = csv.writer(csvfile)
    # Line below only has to be added once, so can be turned off after first run.
    #writer.writerow(fields)
    writer.writerow(rows)
```
